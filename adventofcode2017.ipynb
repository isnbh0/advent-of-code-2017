```python
def get_repeat_sum(num):
    numlist = str(num) + str(num)[0]
    sum_ = 0
    for idx, digit in enumerate(numlist[:-1]):
        sum_ += int(digit == numlist[idx+1])*int(digit)
    return sum_
```

```python
def get_repeat_sum2(num):
    numlen = len(num)
    assert not numlen & 1
    
    numlist = str(num) + str(num)[0]
    
    sum_ = 0
    for idx, digit in enumerate(numlist[:-1]):
        goal = (idx+numlen//2) % numlen
        sum_ += int(digit == numlist[goal])*int(digit)
    
    return sum_
```

```python
def get_weight(mainlist, name):
    if not mainlist[name]['children']:
        return mainlist[name]['weight']
    else:
        return mainlist[name]['weight'] + sum([get_weight(mainlist, child) for child in mainlist[name]['children']])
```

```python
def balanced(mainlist, name):
    if not mainlist[name]['children']:
        return True
    else:
        weights = [get_weight(mainlist, child) for child in mainlist[name]['children']]
        return weights.count(weights[0]) == len(weights), weights
```

```python
def where_to_go(mainlist, name):
    childlist = mainlist[name]['children']
    if not childlist:
        return None
    else:
        is_balanced, weights = balanced(mainlist, name)
        if is_balanced:
            return []
        elif len(weights) == 2:
            return childlist
        else:
            for idx, item in enumerate(weights):
                if weights.count(item) == 1:
                    return [childlist[weights.index(item)]]
```

```python
def simplify_coords(i, j, k):
    # i : n
    # j : se
    # k : sw
    
    # i + j + k -> 0
    while max(min(i, j), min(j, k), min(k, i)) > 0:
        i, j, k = i-1, j-1, k-1
    
    while min(max(i, j), max(j, k), max(k, i)) < 0:
        i, j, k = i+1, j+1, k+1

    return abs(i) + abs(j) + abs(k)
```

```python
def make_coords(dirs):
    max_d = 0
    i, j, k = 0, 0, 0
    n, s, se, nw, sw, ne = 0, 0, 0, 0, 0, 0
    
    for d in dirs:
        if d == 'n':
            i += 1
            n += 1
        elif d == 's':
            i -= 1
            s += 1
        elif d == 'se':
            j += 1
            se += 1
        elif d == 'nw':
            j -= 1
            nw += 1
        elif d == 'sw':
            k += 1
            sw += 1
        elif d == 'ne':
            k -= 1
            ne += 1
        
        cur_d = simplify_coords(i, j, k)
        if cur_d > max_d:
            max_d = cur_d
    
    return i, j, k, n, s, se, nw, sw, ne, max_d
```

```python
a0 = 634
b0 = 301
afac = 16807
bfac = 48271
n = 2**31 - 1
```

```python
def bit16(int0):
    return '{:0>32}'.format(str(bin(int0))[2:])[-16:]
```

15-1

```python
%%time
count = 0

a = a0
b = b0

for it in range(400000):
    a = a*afac % n
    b = b*bfac % n
    #print(a, b)
    if bit16(a) == bit16(b):
        count += 1
print(count)
```
